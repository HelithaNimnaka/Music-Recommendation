## Creating Graph Database

```python
from neo4j import GraphDatabase

# Connect to Neo4j
uri = "bolt://localhost:<PORT>"  # Replace <PORT> with your Neo4j port
user = "<username>"  # Replace with your Neo4j username
password = "<password>"  # Replace with your Neo4j password

driver = GraphDatabase.driver(uri, auth=(user, password))

def create_bidirectional(tx, row):
    tx.run("""
        MERGE (s:Song {name: $track_name})
        SET s.release_date = $release_date,
            s.lyrics = $lyrics,
            s.len = $len,
            s.dating = $dating,
            s.violence = $violence,
            s.world_life = $world_life,
            s.night_time = $night_time,
            s.shake_audience = $shake_audience,
            s.family_gospel = $family_gospel,
            s.romantic = $romantic,
            s.communication = $communication,
            s.obscene = $obscene,
            s.music = $music,
            s.movement_places = $movement_places,
            s.light_visual = $light_visual,
            s.family_spiritual = $family_spiritual,
            s.like_girls = $like_girls,
            s.sadness = $sadness,
            s.feelings = $feelings,
            s.danceability = $danceability,
            s.loudness = $loudness,
            s.acousticness = $acousticness,
            s.instrumentalness = $instrumentalness,
            s.valence = $valence,
            s.energy = $energy,
            s.age = $age

        MERGE (a:Artist {name: $artist_name})
        MERGE (g:Genre {name: $genre})
        MERGE (t:Topic {name: $topic})

        // Bidirectional relationships with Song
        MERGE (s)-[:PERFORMED_BY]->(a)
        MERGE (a)-[:PERFORMED]->(s)
        MERGE (s)-[:HAS_GENRE]->(g)
        MERGE (g)-[:GENRE_OF]->(s)
        MERGE (s)-[:HAS_TOPIC]->(t)
        MERGE (t)-[:TOPIC_OF]->(s)
        """,
        artist_name=row["artist_name"],
        track_name=row["track_name"],
        release_date=int(row["release_date"]),
        lyrics=row["lyrics"],
        len=int(row["len"]),
        dating=float(row["dating"]),
        violence=float(row["violence"]),
        world_life=float(row["world/life"]),
        night_time=float(row["night/time"]),
        shake_audience=float(row["shake the audience"]),
        family_gospel=float(row["family/gospel"]),
        romantic=float(row["romantic"]),
        communication=float(row["communication"]),
        obscene=float(row["obscene"]),
        music=float(row["music"]),
        movement_places=float(row["movement/places"]),
        light_visual=float(row["light/visual perceptions"]),
        family_spiritual=float(row["family/spiritual"]),
        like_girls=float(row["like/girls"]),
        sadness=float(row["sadness"]),
        feelings=float(row["feelings"]),
        danceability=float(row["danceability"]),
        loudness=float(row["loudness"]),
        acousticness=float(row["acousticness"]),
        instrumentalness=float(row["instrumentalness"]),
        valence=float(row["valence"]),
        energy=float(row["energy"]),
        genre=row["genre"],
        topic=row["topic"],
        age=float(row["age"])
    )

with driver.session() as session:
    for idx, row in data.iterrows():
        session.write_transaction(create_bidirectional, row)
        if idx % 1000 == 0:
            print(f"{idx} records processed.")

driver.close()
```

### Recommending songs based on relationships (Graph DB)

```python
def recommend_by_genre(song_name):
    with driver.session() as session:
        result = session.run("""
            MATCH (s:Song {name: $song_name})-[:HAS_GENRE]->(g:Genre)<-[:HAS_GENRE]-(rec:Song)
            WHERE rec.name <> $song_name
            RETURN rec.name AS recommended_song, collect(g.name) AS shared_genres
            LIMIT 10
        """, song_name=song_name)
        return [record for record in result]

# Example:
for rec in recommend_by_genre("i believe"):
    print(rec["recommended_song"], rec["shared_genres"])
```

## Creating Vector Database

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# Choose a transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Use lyrics for semantic embedding, or combine more fields for richer meaning
data['embedding_input'] = data['lyrics'].fillna('')

# Generate embeddings for all songs
embeddings = model.encode(data['embedding_input'].tolist(), show_progress_bar=True)
embeddings = np.array(embeddings)




d = embeddings.shape[1]  # dimension of vectors
index = faiss.IndexFlatL2(d)  # L2 distance for similarity search

index.add(embeddings)  # Add all song vectors

faiss.write_index(index, "songs.index") #Save the index to disk
```

```python
# Reload index
index = faiss.read_index("songs.index")
```

### Recommending songs based on semantic meaning (Vector DB)

```python
def vector_search(query, data, model, index, top_n=10):
    # Convert query to embedding
    query_vec = model.encode([query])
    # Search the index for most similar vectors
    D, I = index.search(np.array(query_vec), top_n)
    # I[0] gives indices of top_n results
    return data.iloc[I[0]]

# Example:
results = vector_search("I feel happy and want energetic pop songs", data, model, index)
print(results[['track_name', 'artist_name', 'genre']])
```
